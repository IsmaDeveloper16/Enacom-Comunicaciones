Creé este archivo para cubrir algunos errores que no habia tenido en cuenta.

importamos las librerias

```python
import pandas as pd
import numpy as np
```

importamos el archivo al cual queremos modificar

```python
tec_provincias = pd.read_csv('nuevos_datos/tecnologia_provincias.csv')
```

creamos una copia

```python
tec_copy = tec_provincias.copy()
```

eliminamos duplicados basandonos en año

```python
tec_copy.drop_duplicates(subset='Año')
```

como vemos en el indice 312 hay un error de carga, el cual pone un * al final, lo que hare sera cambiar eso por el año correspondiente

```python
tec_provincias[tec_provincias['Año'] == '2019 *'] = 2019
```

ahora haremos lo mismo con Trimestre

```python
tec_copy.drop_duplicates(subset='Trimestre')
```

Aqui vemos que en el indice 312, 336 y 360 nos muestra los trimestres mal cargados, lo que hare sera cambiarlos como hice con año.

antes de eso convertiremos la columna en un str

```python
tec_provincias['Trimestre'] = tec_provincias['Trimestre'].astype(str)
```

filtramos en busca de los errores

```python
tec_provincias[(tec_provincias['Trimestre'] != '1') & (tec_provincias['Trimestre'] != '2') & (tec_provincias['Trimestre'] != '3') & (tec_provincias['Trimestre'] != '4')]
```

aqui vemos una serie de errores de carga. lo que hare sera eliminar los errores atravez del indice

obtenemos el indice

```python
indices = tec_provincias[tec_provincias['Trimestre'] == '2019'].index
```

creamos un for para eliminarlos

```python
for i in indices:
    tec_provincias.drop(index=i,inplace=True)
```

cambiamos el tipo de dato

```python
tec_provincias['Trimestre'] = tec_provincias['Trimestre'].astype(int)
```

y cargamos nuevamente los archivos

```python
tec_provincias.to_csv('nuevos_datos/tecnologia_provincias.csv',index=False)
```
